## Scrape the list of third party drivers license locations


**Scrape the list of from [here](https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations), but also include the link. Save as a CSV. Since it's more than just the text from the table, this requires actually using BeautifulSoup**

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36" 
}

response = requests.get('https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations', headers=headers)
doc = BeautifulSoup(response.text)
```

```python
entries = doc.find_all('td')
```

```python
rows = []

for entry in range(4, len(entries), 4):
    row = {}

    soup = BeautifulSoup(str(entries[entry]), "html.parser")
    a_tag = soup.find("a")

    if a_tag and a_tag.has_attr("href"):
        row['company'] = a_tag.text.strip()
        row['company_url'] = a_tag['href']
    else:
        row['company'] = soup.text.strip()
        row['company_url'] = None

    row['address'] = entries[entry + 1].get_text(strip=True)
    row['phone'] = entries[entry + 2].get_text(strip=True)
    row['hours'] = entries[entry + 3].get_text(strip=True)

    rows.append(row)
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv('drivers_license.csv', index=False)
```
